```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Book1.csv')
```

```python
df.head()
```

```python
h=list(df)
```

```python
h
```

```python
data=[]
for i in df.iterrows():
    data.append({
            h[0]:i[1][h[0]],
            h[1]:i[1][h[1]][:-1],
            h[2]:i[1][h[2]][:-1],
            h[4]:i[1][h[4]][:-1],
            h[5]:i[1][h[5]][:-1],
            h[6]:i[1][h[6]][:-1],
        })
```

```python
data
```

```python
import json
with open('data.json', 'w') as outfile:
    json.dump(data, outfile)
```
